#Question 5.

Сколько в процентах было совершено вылетов в Хьюстон из EWR ?

```python

fly_to_hou=rt[(rt.dest== 'IAH')]
all_fly=len(fly_to_hou)
ewr_fly=fly_to_hou.groupby([fly_to_hou.origin== 'EWR']).size()
print((ewr_fly/all_fly)*100)
```

#Question 6. 

Зависимость времени года (месяцев) от задержки прилета.
Здесь я брал среднее значение времени задержки прилета за месяц

```python
import matplotlib.pyplot as plt
group_data=pd.DataFrame(rt.groupby([rt.month])['arr_delay'].mean().reset_index(name='Arr_Delay'))
plt.rcdefaults()
plt.plot(group_data.month,group_data.Arr_Delay,'')
plt.title('Mean Arrival Delay for a Month')
plt.xlim(1,12)
plt.xlabel('Months')
plt.ylabel('Mean_Arr_Delay')
plt.show()
```

Как видно из графика самые большие (по времени) задержки прилетов
происходили в летний период и в декабре. С другой стороны самые лучшие
показатели были сделаны осенью( в том числе в сентябре самолеты прилетали даже раньше).

Какие авиакомпании чаще всего организовывали полет ?

#Question 8.

Узнаем какая из авиакомпаний имеет самый большой процент задержки (в отправке) рейсов

```python
fly_dep_delay_carrier=rt[rt.dep_delay>0]
group_data1=pd.DataFrame(rt.groupby([rt.carrier]).size().reset_index(name='Total_Count'))
group_data2=pd.DataFrame(fly_dep_delay_carrier.groupby([fly_dep_delay_carrier.carrier]).size().reset_index(name='Count'))
rt_merged = pd.merge(group_data1, group_data2, on=['carrier'])
rt_merged['Percent_Delay'] = (rt_merged.Count/rt_merged.Total_Count)*100
ind = np.arange(len(rt_merged.carrier))
plt.rcdefaults()
plt.bar(ind, rt_merged.Percent_Delay, color="green")
plt.xlabel('Carriers')
plt.ylabel('Percent Delay')
plt.title('Percent Delay by Carrier')
plt.xticks(ind, rt_merged.carrier)
plt.show()
```

Как видно из диаграммы, чаще всего задержки при вылете были у авиакомпаний WIN, FL и F9. Реже всего - у US и HA.

#Question 10.

Заисимость между температурой и задержками прилетов

```python
fly_arr_delay=rt[(rt.arr_delay>0)]
group_data1=fly_arr_delay.groupby([fly_arr_delay.origin,fly_arr_delay.month,fly_arr_delay.day,fly_arr_delay.hour])
group_data2=group_data1['arr_delay'].agg([np.size,np.mean]).reset_index()
wt_merge=pd.merge(group_data2,wt,on=['origin','hour','day','month'])
wt_merge=wt_merge.rename(columns={'size':'Count','mean':'Delay'})
wt_temp=wt_merge.groupby(wt_merge.temp)['Delay','Count'].agg([np.mean]).reset_index()
plt.rcdefaults()
plt.scatter(wt_temp.temp, wt_temp.Delay)
plt.title('Arr_mean_Delay time & Temp')
plt.xlim(10, 100)
plt.ylim(0, 150)
plt.ylabel('Arrival mean Delay Time')
plt.xlabel('Temp')
plt.show()
```

Как видно из диаграммы особой зависимости между температурой воздуха и времени задержки прилета особо нет. Однако, начиная с 70 градусов время задержки слегка увеличивается, ровно также как и от 10 до 45 градусов.
